```python
# export
import os
from code.config import * 
from loguru import logger
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 100)  # 设置显示数据的最大列数，防止出现省略号…，导致数据显示不全
pd.set_option('expand_frame_repr', False)  # 当列太多时不自动换行

from sklearn.metrics import f1_score
```

# load data

```python
train_user_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/train_user_feat2.pkl'))

test_user_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/test_user_feat2.pkl'))

train_voc_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/df_train_voc2.pkl'))

test_voc_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/df_test_voc2.pkl'))

train_sms_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/df_train_sms2.pkl'))

test_sms_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/df_train_sms2.pkl'))

train_app_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/df_train_app2.pkl'))

test_app_feat = pd.read_pickle(os.path.join(args.DATA_DIR,'data_gen/df_train_app2.pkl'))

```

```python
df_train = pd.merge(pd.merge(pd.merge(train_user_feat, train_voc_feat.reset_index(), on='phone_no_m month_id'.split(), how='left'),
                    train_sms_feat.reset_index(), on='phone_no_m month_id'.split(), how='left'),
                   train_app_feat.reset_index(), on='phone_no_m month_id'.split(), how='left').reset_index(drop=True)

df_test = pd.merge(pd.merge(pd.merge(test_user_feat, test_voc_feat.reset_index(), on='phone_no_m month_id'.split(), how='left'),
                    test_sms_feat.reset_index(), on='phone_no_m month_id'.split(), how='left'),
                   test_app_feat.reset_index(), on='phone_no_m month_id'.split(), how='left').reset_index(drop=True)
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
lbe_city = LabelEncoder()
lbe_county = LabelEncoder()
lbe_city.fit(pd.Series(list(set(df_train["city_name"]).union(
    set(df_test["city_name"])))).astype(np.str))
lbe_county.fit(pd.Series(list(set(df_train["county_name"]).union(
    set(df_test["county_name"])))).astype(np.str))
df_train["city_name"] = lbe_city.transform(
    df_train["city_name"].astype(np.str))
df_train["county_name"] = lbe_county.transform(
    df_train["county_name"].astype(np.str))

df_test["city_name"] = lbe_city.transform(
    df_test["city_name"].astype(np.str))
df_test["county_name"] = lbe_county.transform(
    df_test["county_name"].astype(np.str))
```

# train 

```python
from sklearn.model_selection import KFold
folds = KFold(n_splits=5, shuffle=True, random_state=2020)
```

```python
params = {
    'objective': 'binary',
    'metric': 'auc',
    'num_leaves': 31,
    'max_bin': 50,
    'max_depth': 16,
    "learning_rate":0.1,
    "colsample_bytree":0.8,#每次迭代中随机选择特征的比例
    'min_child_samples': 25,
    'verbosity': -1,    #信息输出设置成1则没有信息输出
    'seed':1024,
    }
```

```python
import lightgbm as lgb
```

```python
predictions = np.zeros([len(df_test)])
for fold_,(trn_idx,val_idx) in enumerate(folds.split(df_train)):
    X_train, y_train = df_train.iloc[trn_idx].copy(), df_train.label.iloc[trn_idx].copy()
    X_val, y_val = df_train.drop(columns='label').iloc[val_idx], df_train.label.iloc[val_idx]
    
#     city_name_series = X_train.groupby('city_name')['label'].mean()
#     county_name_series = X_train.groupby('county_name')['label'].mean()
    
    del X_train['label']
    
#     X_train['city_name'] = X_train['city_name'].map(city_name_series, na_action='ignore')


#     X_train['county_name'] = X_train['county_name'].map(county_name_series, na_action='ignore')

#     X_val['city_name'] = X_val['city_name'].map(city_name_series, na_action='ignore')
#     X_val['county_name'] = X_val['county_name'].map(county_name_series, na_action='ignore')
#     df_test['city_name'] = df_test['city_name'].map(city_name_series, na_action='ignore')
#     df_test['county_name'] = df_test['county_name'].map(county_name_series, na_action='ignore')
    
    
    trn_data = lgb.Dataset(X_train, y_train)
    val_data = lgb.Dataset(X_val, y_val)
    
    num_round = 1000
    # callback
    def reset_metrics():
        def callback(env):
            if env.iteration % 100 == 1:
                rs = pd.Series(env.model.predict(X_val).tolist()).map(round)
                print(f'{env.iteration} predict...{f1_score(y_val.tolist(), rs.tolist(), average="macro")}')
#                 print(f'{env.iteration} predict...{accuracy_score(y_val.values, rs.values)}')
        callback.before_iteration = False
        callback.order = 0
        return callback
    
    clf = lgb.train(param, 
                    trn_data, 
                    num_round, 
                    valid_sets = [trn_data, val_data], 
                    verbose_eval = 100, 
                    early_stopping_rounds = 200,
                    callbacks=[reset_metrics()])
    rs = pd.Series(clf.predict(X_val).tolist()).map(round)
    print(f'{num_round} predict...{f1_score(y_val.tolist(), rs.tolist(), average="macro")}')
    
    importance = clf.feature_importance()
    feature_name = clf.feature_name()
    feature_importance = pd.DataFrame({'feature_name':feature_name,'importance':importance} ).sort_values(by='importance',ascending=False)
#     feature_importance.to_csv('feature_importance_{}.csv'.format(idx+1),index=None)
    print(feature_importance.head(10))
    #oof[val_idx] = clf.predict(X_train[val_idx], num_iteration=clf.best_iteration)    
    predictions += clf.predict(df_test, num_iteration=clf.best_iteration) / folds.n_splits
```
